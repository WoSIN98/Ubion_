```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
```

```python
url = 'https://www.coupang.com/np/categories/194276'
driver = webdriver.Chrome()
driver.get(url)
```

```python
productList = driver.find_element(By.ID, 'productList')
lis = productList.find_elements(By.TAG_NAME, 'li')
```

```python
prd_name = []
prd_price = []
prd_deli = []
prd_url = []
for idx, li in enumerate(lis):
    try:
        prd_name.append(li.find_element(By.CLASS_NAME, 'name').text)
        prd_price.append(li.find_element(By.CLASS_NAME, 'price-value').text)
        prd_deli.append(li.find_element(By.CLASS_NAME, 'delivery').text)
        prd_url.append(li.find_element(By.CLASS_NAME, 'baby-product-link').get_attribute('href'))

        e = li.find_element(By.TAG_NAME, 'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()

        food_name = './image/cupang/food{}.jpg'.format(idx+1)
        with open(food_name, 'wb') as f:
            f.write(img_res.content)


        if idx >= 5:
            break
    except Exception:
        pass
        
    
```

```python
url = 'https://www.e-himart.co.kr/app/display/showDisplayCategory?dispNo=1011030000#pageCount=1'
driver = webdriver.Chrome()
driver.get(url)

productList = driver.find_element(By.ID, 'productList')
lis = productList.find_elements(By.TAG_NAME, 'li')

prd_name = []
prd_price = []
prd_beneprice = []
prd_star = []
prd_url = []
for idx, li in enumerate(lis):
    try:
        prd_name.append(li.find_element(By.CLASS_NAME, 'prdName').text)

        prd_price.append(li.find_elements(By.CLASS_NAME, 'discountPrice')[0].text)

        tmpDiscountPrice = li.find_elements(By.CLASS_NAME, 'discountPrice')
        if (len(tmpDiscountPrice) == 2):
            prd_beneprice.append(li.find_elements(By.CLASS_NAME, 'discountPrice')[1].text)
        else:
            prd_beneprice.append(li.find_elements(By.CLASS_NAME, 'discountPrice')[0].text)

        prd_star.append(li.find_element(By.CLASS_NAME, 'ratingPoint').text)
        prd_url.append(li.find_element(By.CLASS_NAME, 'prdLink').get_attribute('href'))

        e = li.find_element(By.TAG_NAME, 'img')
        img_url = e.get_attribute('src')
        img_res = requests.get(img_url)
        img_res.raise_for_status()

        food_name = './image/himart/세탁기{}.jpg'.format(idx+1)
        with open(food_name, 'wb') as f:
            f.write(img_res.content)


        if idx >= 5:
            break
    except Exception:
        pass
```

```python

final = pd.DataFrame({'이름' : prd_name, '가격' : prd_price, '최대혜택가': prd_beneprice,'평점' : prd_star, 'URL' : prd_url})
final.to_csv('하이마트세탁기.csv')
final
```
