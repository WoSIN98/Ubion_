### IFrame popup

```python
df = pd.read_excel('./서울지역 대학교 위치.xlsx')
df.rename(columns = {
    'Unnamed: 0' : '대학교'},
                 inplace = True
)
                
seoul_map = folium.Map(location = [37.55,126.98], zoom_start = 12)
                 
```

```python
from folium.map import Popup
import pandas as pd
import folium
```

```python
for name, lat, lng in zip(df.대학교, df.위도, df.경도):
    iframe = folium.IFrame(name, width = 300, height = 80)              # IFrame에 텍스트 입력
    popup = folium.Popup(iframe, max_width = 300)                       # Popup에 IFrame 밀어넣기
    folium.Marker([lat, lng], popup = popup).add_to(seoul_map) 

seoul_map.save('./seoul_colleges.html')
seoul_map
```

### 원형 마커

```python
df = pd.read_excel('./서울지역 대학교 위치.xlsx')
df.rename(columns = {
    'Unnamed: 0' : '대학교'},
                 inplace = True
)
                
seoul_map = folium.Map(location = [37.55,126.98], zoom_start = 12)
                 
```

```python
for name, lat, lng in zip(df.대학교, df.위도, df.경도):
    folium.CircleMarker([lat, lng],
                       radius = 10,                 # 원의 반지름
                       color = 'blue',             # 원의 둘레 색상
                       fill = True,               
                       fill_color = 'skyblue',        # 원을 채우는 색
                       fill_opacity = 0.7           # 투명도
        ).add_to(seoul_map)
    
seoul_map.save('./seoul_colleges2.html')
seoul_map
```
