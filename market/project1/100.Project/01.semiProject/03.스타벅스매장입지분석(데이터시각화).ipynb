#### 01-2. 시군구별 스타벅스 매장 수 시각화

```python
# 서울시 시군구별 통계 데이터 불러오기
seoul_sgg_stat = pd.read_excel('./file/seoul_sgg_stat.xlsx')
seoul_sgg_stat.head(3)
```

```python
# 서울시 시군구 행정 경계 지도 파일 불러오기
sgg_geojson_file_path = './maps/seoul_sgg.geojson'
seoul_sgg_geo = json.load(open(sgg_geojson_file_path, encoding='utf-8'))
print(type(seoul_sgg_geo))
seoul_sgg_geo
```

```python
print(type(seoul_sgg_geo['features']))
seoul_sgg_geo['features']
```

```python
seoul_sgg_geo['features'][0]['properties']
# SIG_CD : 시군구코드
# SIG_KOR_NM : 시군구명
```

```python
starbucks_bubble = folium.Map(
    location = [37.573050, 126.979189], tiles= 'cartoDB dark_matter', zoom_start = 11)
```

```python
starbucks_bubble
```

```python
# 서울시 시군구 경계 지도 그리기
def style_function(feature):                 # 변수 만들어 주는 fuc : 콜백 fuc
    return {
        'opacity': 0.7,
        'weight' : 1,
        'color' : 'white',
        'fillOpacity' : 0,
        'dashArray' : '5, 5'
    }
folium.GeoJson(
    seoul_sgg_geo,
    style_function= style_function
).add_to(starbucks_bubble)
```

```python
starbucks_bubble
```

```python
# 서울시 시군구별 스타벅스 평균 매장 수 계산
starbucks_mean = seoul_sgg_stat['스타벅스_매장수'].mean()
starbucks_mean
```

```python
## 서울시 시군구별 스타벅스 매장 수를 버블 지도로 시각화
# 결과 : 빨간색으로 표시된 버블들은 서울시 평균 스타벅스 매장 수보다 많은 매장이 위치한 시군구
# 강남구, 서초구, 송파구, 중구, 종로구, 영등포구, 마포구, 서대문구 등에 스타벅스 매장이 많이 입점해있다.

for idx in seoul_sgg_stat.index:
    lat = seoul_sgg_stat.loc[idx, '위도']
    lng = seoul_sgg_stat.loc[idx, '경도']
    count = seoul_sgg_stat.loc[idx, '매장개수']
    
    if count > starbucks_mean:
        fillColor = 'red'
    else:
        fillColor = 'blue'
        
    folium.CircleMarker(
        location=[lat, lng], 
        color='#FFFF00',
        fill_color=fillColor, 
        fill_opacity=0.7,
        weight=1.5,
        radius=count/2
    ).add_to(starbucks_bubble)
    
starbucks_bubble
```

```python
## 서울시 시군구별 스타벅스 매장 수를 단계 구분도로 시각화

sgg_geojson_file_path = './maps/seoul_sgg.geojson'
seoul_sgg_geo_2 = json.load(open(sgg_geojson_file_path, encoding= 'utf-8'))
starbucks_choropleth = folium.Map(
    location = [37.573050, 126.979189],
    tiles= 'cartoDB dark_matter', zoom_start = 11
)

folium.Choropleth(
    geo_data= seoul_sgg_geo_2,
    data = seoul_sgg_stat,
    columns = ['시군구명','스타벅스_매장수'],
    fill_color = 'YlGn',
    fill_opacity=0.7,
    line_opacity=0.5,
    key_on='properties.SIG_KOR_NM'
).add_to(starbucks_choropleth)
```

```python
starbucks_choropleth
```

### 02. 가설검증

#### 02-1. 가설1 : "거주인구가 많은 지역에 스타벅스 매장이 많이 입지해 있을 것이다."
- 스타벅스 매장 수와 인구수 비교

```python
# 시군구별 통계 데이터
seoul_sgg_stat = pd.read_excel('./file/seoul_sgg_stat.xlsx')
seoul_sgg_stat.head(3)
```

```python
# 서울시 시군구 행정 경계 지도
sgg_geojson_file_path = './maps/seoul_sgg.geojson'
seoul_sgg_geo_2 = json.load(open(sgg_geojson_file_path, encoding = 'utf-8'))
```

```python
## 서울시 시군구별 주민등록 인구수 단계구분도 지도 시각화
starbucks_choropleth = folium.Map(
    location=[37.573050, 126.979189],
    tiles = 'cartoDB dark_matter', zoom_start=11
)

folium.Choropleth(
    geo_data= seoul_sgg_geo_2,
    data = seoul_sgg_stat,
    columns = ['시군구명', '주민등록인구'],
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.5,
    key_on='properties.SIG_KOR_NM'
).add_to(starbucks_choropleth)
```

```python
starbucks_choropleth
# 위의 그래프와 비교하면 주민등록인구수와 스타벅스 매장수는 비례하지 않는다는 것을 알 수 있다. 시각화하여 비교 평가하면 설득력 향상.
```

```python
## 인구 만명당 스타벅스 매장 수 컬럼 추가
seoul_sgg_stat['만명당_매장수'] = (seoul_sgg_stat['스타벅스_매장수'] * 10000) / seoul_sgg_stat['주민등록인구']
```

```python
seoul_sgg_stat.head()
```

```python
## 인구 만 명당 스타벅스 매장 수 지도 시각화
sgg_geojson_file_path = './maps/seoul_sgg.geojson'
seoul_sgg_geo_2 = json.load(open(sgg_geojson_file_path, encoding = 'utf-8'))

viz_map_1 = folium.Map(
    location=[37.573050, 126.979189],
    tiles = 'cartoDB dark_matter', zoom_start=11
)

def style_function(feature):
    return{
        'opacity' : 0.7,
        'weight' : 1,
        'fillOpacity' : 0,
    }

folium.GeoJson(
    seoul_sgg_geo_2,
    style_function= style_function
).add_to(viz_map_1)
```

```python
viz_map_1
```

```python
# 만명당 매장수 기준 상위 10% 추출 값
top = seoul_sgg_stat['만명당_매장수'].quantile(0.9)
for idx in seoul_sgg_stat.index:
    lat = seoul_sgg_stat.loc[idx, '위도']
    lng = seoul_sgg_stat.loc[idx, '경도']
    r = seoul_sgg_stat.loc[idx, '만명당_매장수']

    if r > top: 
        fillColor = '#FF3300' # (Red)
    else:
        fillColor = '#CCFF33' # (Green)

    folium.CircleMarker(
        location=[lat, lng], 
        color='#FFFF00',      # (Yellow)
        fill_color=fillColor, 
        fill_opacity=0.7,
        weight=1.5,
        radius= r*10
    ).add_to(viz_map_1)
```

```python
viz_map_1
```

```python
## 결론 
# 시군구별 인구 단계 구분도에서 중구와 종로구가 서울에서 주민등록인구가 적은 지역으로 표시되었으나, 인구 1만 명당 스타벅스 매장수는 가장 많은 것으로 표시
# 가설 1("거주 인구가 많은 지역에 스타벅스 매장이 많이 입지해 있을 것이다")은 성립되지 않음.
```

#### 02-2. 사설 2 : "직장인이 많은 지역에 스타벅스 매장이 많이 입지해 있을 것이다."
- 스타벅스 매장 수와 사업체 종사자 수

```python
# 종사자수 만 명당 스타벅스 매장 수 컬럼 추가
seoul_sgg_stat['종사자수_만명당_매장수'] = (seoul_sgg_stat['스타벅스_매장수']*10000) / seoul_sgg_stat['종사자수']
seoul_sgg_stat.head(3)
```

```python
## 종사자수 만 명당 스타벅스 매장 수 지도 시각화
sgg_geojson_file_path = './maps/seoul_sgg.geojson'
seoul_sgg_geo_1 = json.load(open(sgg_geojson_file_path, encoding = 'utf-8'))

viz_map_1 = folium.Map(
    location=[37.573050, 126.979189],
    tiles = 'cartoDB dark_matter', zoom_start=11
)

folium.GeoJson(
    seoul_sgg_geo_1,
    style_function= style_function
).add_to(viz_map_1)
```

```python
top = seoul_sgg_stat['종사자수_만명당_매장수'].quantile(0.9)
for idx in seoul_sgg_stat.index:
    name = seoul_sgg_stat.at[idx, '시군구명']
    lat = seoul_sgg_stat.loc[idx, '위도']
    lng = seoul_sgg_stat.loc[idx, '경도']
    r = seoul_sgg_stat.loc[idx, '종사자수_만명당_매장수']
    
    if r > top:
        fillColor = '#FF3300'
    else:
        fillColor = '#CCFF33'
    
    folium.CircleMarker(
        location=[lat, lng], 
        color='#FFFF00', 
        fill_color=fillColor, 
        fill_opacity=0.7,
        weight=1.5,
        radius= r * 10
    ).add_to(viz_map_1)

viz_map_1
```
