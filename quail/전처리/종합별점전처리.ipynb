```python
import pandas as pd
```

```python
review_df = pd.read_csv('../data2/review_df.csv')
```

```python
review_df.head(2)
```

```python
review_rating = review_df[['merchant', 'user_name','user_rank','num_response','user_star','rating']]
```

```python
(review_rating['rating'] - review_rating['user_star']).mean()
```

```python
((review_rating['rating'] - review_rating['user_star']) * review_rating['user_rank']).mean()
```

```python
review_rating
```

```python
rank_score = []
for rating in review_rating['user_rank']:
    if 0 < rating <= 10 :
        rank_score.append(1)
    elif 10 < rating <= 30 :
        rank_score.append(1.1)
    elif 30 < rating <= 60 :
        rank_score.append(1.2)
    elif 60 < rating <= 80 :
        rank_score.append(1.3)
    else:
        rank_score.append(1.5)
```

```python
review_rating['rank_score'] = rank_score
```

```python
review_rating
```

```python
review_rating['total_rating'] = (review_rating['rating'] - review_rating['user_star'])*review_rating['rank_score']
```

```python
review_rating
```

```python
total_rating = review_rating[['merchant','total_rating']].groupby('merchant').mean()
```

```python
total_rating
```

```python
total_rating.mean()
```

```python
# 데이터 저장
total_rating.to_csv('../data2/total_rating.csv')
```
