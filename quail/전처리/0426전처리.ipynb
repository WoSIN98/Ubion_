```python
import os
import pandas as pd

### 225개의 음식점 데이터를 하나의 데이터프레임으로 변환. (리뷰가 없는 27개 데이터는 포함되지 않는다.)

# 파일이 위치한 디렉토리 경로
directory_path = '../data2/최종 data2'

# 모든 CSV 파일을 읽고 파일명을 새로운 열로 추가
all_data = []
for file in os.listdir(directory_path):
    if file.endswith('.csv'):
        file_path = os.path.join(directory_path, file)
        df = pd.read_csv(file_path)
        df['merchant'] = file.strip('.csv')  # 파일명을 새로운 열로 추가
        all_data.append(df)

# 모든 데이터를 하나의 데이터프레임으로 결합
combined_df = pd.concat(all_data, ignore_index=True)

# 결과 확인
combined_df
```

```python
len(combined_df['merchant'].unique())
```

```python
combined_df.info()
```

```python
## merchant열을 가장 앞으로 옮기기 
# 'merchant' 열을 먼저 추출하고 데이터프레임의 첫 번째 위치로 추가
merchant_col = combined_df.pop('merchant')
combined_df.insert(0, 'merchant', merchant_col)

# 결과 확인
print(combined_df.head())
```

```python
combined_df.to_csv('../data2/review_df.csv',index=False)
```
