```python
## api에 요청을 보낸다. 
# 특정 주소에 요청 -> 해당하는 서버에서 응답 메시지를 보내준다.
response = requests.get(url, params = _param)                             # (주소, 요청변수)   # 이건 블록 따로 처리
```
