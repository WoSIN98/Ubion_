### 복습
1. response3 데이터를 문자열로 저장
2. 문자열로 저장된 데이터를 BeautifulSoup를 이용하여 parsing
3. div 중 id가 content인 태그를 추출
4. 3번에서 추출한 Tag에서 div 중 class가 today_section인 태그를 추출
5. 헤드라인만 추출하여 리스트로 생성

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url3 = 'https://sports.news.naver.com/index'
response3 = requests.get(url3)
```

```python
html_data3 = response3.text
html_data3
```

```python
soup3 = bs(html_data3, 'html.parser')
soup3
```

```python
len(soup3.find_all('div', attrs={'class' : 'today_section'}))
```

```python
div_data = soup3.find('div', attrs={'class' : 'today_section'})
div_data
```

```python
div_data.find_all('strong')
```

```python
head_data = div_data.find_all('strong')
headline = list()
for i in head_data:
   headline.append(i.get_text())

print(headline)
```
