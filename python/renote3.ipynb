```python
def mean(*ags):
    sum = 0
    num = 0
    for i in ags:
        sum += i 
        num += 1
    res = sum / num 
    return res
```

```python

def c_min(ags):
    res = ags[0]
    for i in range(1, len(ags),1):
        if res > ags[i]:
            res = ags[i]
    return res   
```

```python
def c_sort(*ags):
    result = []
    ags = list(ags)
    while True:
        if len(ags) == 0: 
            break
        min_ags = c_min(ags)
        result.append(min_ags)
        ags.remove(min_ags)
    return result
```

```python
c_sort(3,9,8,2,5)
```

```python
def c_sort(*ags, reverse = True):
    result = []
    ags = list(ags)
    while ags:
        if reverse:
            ags_ = max(ags)
        else:
            ags_ = c_min(ags)
        
        result.append(ags_)

        ags.remove(ags_)
    
    return result
```

```python
c_sort(3,9,8,2,5 , reverse = False)
```

```python
c_sort(3,9,8,2,5 , reverse = True)
```

```python
class Calculate:
    def __init__(self, input_x, input_y):
        self.x = input_x
        self.y = input_y
    def add(self):
        result = self.x + self.y
        return result
    def subtract(self):
        result = self.x - self.y
        return result
```

```python
c_class = Calculate(10,5)
```

```python
c_class.subtract()
```

```python
class Calculate2(Calculate):
    def __init__(self, input_x, input_y):
        super().__init__(input_x, input_y)
        
    def mul(self):
        result = self.x * self.y
        return result 
```

```python
c_class = Calculate2(10, 5)
```

```python
c_class.mul()
```
