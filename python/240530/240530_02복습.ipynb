### 복습
1. csv 폴더 안에 있는 superstore.xlsx 파일을 로드
    - orders 시트 로드
    - returns 시트 로드
2. 1번의 과정에서 로든한 두 개의 데이터프레임을 조인 결합(left)
3. Row ID를 인덱스로 변경 
4. 컬럼의 이름을 변경
    - [주문번호, 주문일자, 배송일자, 배송종류, 고객번호, 고객이름, 고객종류, 국가, 지역, 시/도, 시/군/구, 제품번호, 제품명, 분류1, 분류2, 매출, 개수, 할인율, 수익, 반품여부]
5. 새로운 파생변수 생성(수익율)
    - 수익 / 매출 * 100
6. csv 파일로 저장 (superstore.csv)

```python
total = pd.merge(orders, returns, how='left', on = 'Order ID')
```

```python
total.head()
```

```python
total.set_index('Row ID', inplace=True)
```

```python
total.columns = ['주문번호', '주문일자', '배송일자', '배송종류', '고객번호', '고객이름', '고객종류', '국가', '지역', '시/도', '시/군/구', '제품번호', '제품명', '분류1', '분류2', '매출', '개수', '할인율', '수익', '반품여부']
```

```python
total.head()
```

```python
# 5. 새로운 파생변수 생성(수익율)   - 수익 / 매출 * 100
total['수익률'] = total['수익'] / total['매출'] * 100
```

```python
total.to_csv('./data/superstore.csv',index=False)
```
