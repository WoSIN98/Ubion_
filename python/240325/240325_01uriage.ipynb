#### lambda 함수 
- 일반함수는 선언과 호출 2가지의 과정으로 나누어 진다.
- lambda 함수는 선언과 동시에 호출이 자동으로 된다.
    - lambda 매개변수명 : 실행이 될 코드 
- 일반함수로 사용을 하기에는 너무 간단한 작업은 lambda로 생성

```python
## case2 ( map() -> 1차원 자료형 구조 데이터에서 각 데이터들을 하나씩 추출하여 작업(함수) )                            # 많이 사용한다.

# Base(python) : map( 함수, 1차원 리스트명 )
_list = [1,2,3,4]
# map(), lambda 결합
list(map(
    lambda x : x + 1,
    _list
))            # [2,3,4,5] 결과값
# map(), 일반 함수 결합
def change(x):
    return x + 2
list(                                   # 반복문을 간단하게 사용할 수 있다.
    map(
        change,                          # 함수명만 쓰는 것. 인자값 적지 않는다
        _list
    )
)
```

- 결측치를 다른 값으로 대체 
    1. 상품명에 따라 상품 가격들이 고정 
    2. 결측치를 상품명에 맞춰서 해당 상품의 결측치가 아닌 상품 가격을 결측치에 대체
        - ex) 상품A 인 데이터 중 item_price가 결측치가 아닌 데이터의 평균 값을 구하고, 해당 평균 값을 상품A인 데이터 중 item_price가 결측치인 부분에 대입 

```python
## 상품A인 데이터만 출력 
# 인덱스 조건식(item_name == '상품A')
flag = df2['item_name'] == '상품A' 
df2.loc[flag]
```

```python
import numpy as np
```

```python
## item_price가 결측치인 데이터만 출력

# 인덱트 조건식(item_price가 결측치인)
# 결측치는 연산이 되지 않기때문에 비교연산자를 사용해도 False 출력
# df2['item_price'] == np.nan                이러면 결측치 == 결측치 -> 결과값: False
# 결측치의 유무를 판단하는 isna() 함수를 이용
flag_null = df2['item_price'].isna() 
df2.loc[flag_null]
```

```python
## item_name이 상품A이고 item_price가 결측치인 데이터를 출력
# 인덱스 조건식(item_name == '상품A' (flag) and item_price가 결측치인 (flag_null))
df2.loc[
    flag & flag_null
]
```

```python
## item_name이 상품A이고 item_price가 결측치가 아닌 데이터를 출력
# 인덱스 조건식(item_name == '상품A' (flag) and item_price가 결측치가 아닌 (-flag_null))
# flag_null의 부정을 출력
# not(flag_null)              # 에러발생 : 시리즈 형태여서 에러 
# -flag_null                  # -도 된다. 왜 되는 지는 모르겠다.
# 비트연산자(~) : 2진수로 이루어진 데이터에서 0과 1을 반대로 교체 
# 인덱스 조건식에 부정형 연산자로 사용이 가능
print(~flag_null)                    # 비트연산자(~) 뒤집어준다. 비트값을 변환해준다. T,F로 되어있는 집합에서는 부정형으로 사용할 수 있다.
df2.loc[
    flag & ~flag_null
]
```

```python
## 상품 A이고 가격이 결측치가 아닌 item_price의 평균을 구한다. 
mean_data = df2.loc[flag & ~flag_null,'item_price'].mean()
mean_data
```

```python
## mean_data를 상품A이고 가격이 결측치인 데이터에 결측치 대체
# case1 fillna()를 이용하여 결측치에 데이터를 대입
df2.loc[flag & flag_null].fillna(mean_data)
# case2 Series 데이터에 특정 데이터를 대입
df2.loc[flag & flag_null,'item_price'] = mean_data
```

```python
## item_name 데이터들의 고유한 값들을 출력

# 1.함수를 이용 -> 고유하다(unique)
df2['item_name'].unique()
# 2.집합의 형태로 데이터를 변환
# {} -> dict, set(집합)
list(set(df2['item_name']))                                                     # 리스트로 묶어준다.
# 3.빈도수를 체크하는 함수
# 값(value) + 갯수(count) -> value_counts()
df2['item_name'].value_counts().index                               # df2['item_name'].value_counts()는 스리즈 형태 
```

```python
# 4. for문 
# item_name에 중복 데이터를 제거한 새로운 리스트를 출력
# for문을 이용한다.

# 내가 한 코드
name_list = list(df2['item_name'])
# 새로운 빈 리스트(result)를 생성
result = []
# item_name의 값들을 기준으로 반복문 실행 (변수 i)
for i in name_list:
    # reslut에 i가 존재하지 않는가?
    if result.count(i) == 0:
        # 존재하지 않는다면 result에 i를 추가
        result.append(i)
# reslut를 출력
result.sort()
print(result)
```

```python
# 수업
name_list = list(df2['item_name'])
result = []
for i in name_list:
    # case1 (비교연산자 -> 포함(in) 되어있지 않다(not).)
    # if i not in result:
    # case2 (list의 내장함수 이용 : count())
    if result.count(i) == 0:   
        result.append(i)
print(result)
```

```python
name_list = list(df2['item_name'])
result = []
for i in name_list:
    # case3 (list 내장함수 중 index())
    # index() : 특정 데이터의 위치를 출력 -> 데이터가 존재하지 않으면 에러 발생
    try:
        # 시도할 코드 
        result.index(i)
    except:
        # try 구문의 코드가 에러가 발생한다면
        # 예외 처리
        result.append(i)
print(result)
```

```python
item_list = df2['item_name'].unique()

for item in item_list:
    # print(item)                            # 혼자 코드 짤 때 하나 실행 하나 print, 하나 실행 하나 print 확인해나가는 것이 좋다.
    # item : item_list에 원소들이 하나씩 대입 
    # item 변수를 이용하여 인덱스의 조건식 생성
    # 상품A부터 상품Z까지의 조건식을 반복적으로 생성
    flag = df2['item_name'] == item
    # print(flag.head(3))
    # item_price가 결측치인 인덱스 조건식 
    flag_null = df2['item_price'].isna()

    # flag 조건식과 flag_null의 부정 조건식에 item_price의 평균 
    mean_data = df2.loc[flag & ~flag_null, 'item_price'].mean()
    # mean_data를 flag 조건식과 flag_null 조건식에 item_price 대입
    df2.loc[flag & flag_null , 'item_price'] = mean_data

    
```
