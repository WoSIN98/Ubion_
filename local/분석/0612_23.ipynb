```python
import pandas as pd
```

```python
df = pd.read_csv('../data/경기도/전처리 후/회귀df결측치포함.csv')
total = pd.read_csv('../data/경기도/전처리 후/test.csv')
```

```python
df
```

```python
df.sort_values(['시군명','기준년월'],inplace=True)
df
```

```python
df = df.fillna(0)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv('../data/경기도/전처리 후/회귀df결측치포함.csv',index=False)
```

```python
## 인센티브 변화 컬럼 생성
df['인센티브 변화'] = df['인센티브'] - df.groupby('시군명')['인센티브'].shift(1)
df = df.fillna(0)
df
```

```python
df['인센티브 변화'].unique()
```

```python
## 지급한도 변화 컬럼 생성
df['지급한도 변화'] = df['월지급한도'] - df.groupby('시군명')['월지급한도'].shift(1)
df = df.fillna(0)
df
```

```python
## 전달대비 충전액 컬럼 생성
df['전달 대비 충전액'] = df['월별충전액'] - df.groupby('시군명')['월별충전액'].shift(1)
df = df.fillna(0)
df
```

```python
## 전달대비 사용액 컬럼 생성
df['전달 대비 사용액'] = df['월별사용액'] - df.groupby('시군명')['월별사용액'].shift(1)
df = df.fillna(0)
df
```

```python
df.to_csv('../data/경기도/전처리 후/충전액 변화.csv',index=False)
```

```python
import matplotlib.pyplot as plt
from matplotlib.dates import date2num
import seaborn as sns
import numpy as np
# matplotlib 한글 
import platform
if platform.system() == 'Darwin' : 
    plt.rc('font', family = 'AppleGothic')
else:
    plt.rc('font', family = 'Malgun Gothic')
```

```python
from matplotlib.dates import date2num

# Re-load the data to ensure it's clean
data = df

# Convert '기준년월' to datetime and then to numeric format for regression plotting
data['기준년월'] = pd.to_datetime(data['기준년월'])
data['기준년월_numeric'] = date2num(data['기준년월'].dt.to_pydatetime())

# Filter necessary columns
trend_data = data[['기준년월_numeric', '시군명', '인센티브 변화', '지급한도 변화', '전달 대비 충전액', '전달 대비 사용액']]

# Unique regions
regions = trend_data['시군명'].unique()

# Initialize plot setup
fig, axs = plt.subplots(len(regions), 2, figsize=(15, 5 * len(regions)), sharex=True)

# Plot trend lines for each region
for idx, region in enumerate(regions):
    region_data = trend_data[trend_data['시군명'] == region]

    # Plot '전달 대비 충전액' trend
    sns.regplot(x='기준년월_numeric', y='전달 대비 충전액', data=region_data, ax=axs[idx, 0], color='blue', label='전달 대비 충전액')
    axs[idx, 0].set_title(f'{region} - 전달 대비 충전액')
    axs[idx, 0].legend()
    
    # Plot '전달 대비 사용액' trend
    sns.regplot(x='기준년월_numeric', y='전달 대비 사용액', data=region_data, ax=axs[idx, 1], color='green', label='전달 대비 사용액')
    axs[idx, 1].set_title(f'{region} - 전달 대비 사용액')
    axs[idx, 1].legend()

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

```python
import statsmodels.api as sm
# Filter out rows where '인센티브 변화' is 0
new_filtered_data = df[df['인센티브 변화'] != 0]

# Selecting the required columns for the regression model
new_model_data = new_filtered_data[['전달 대비 충전액', '인센티브 변화', '지급한도 변화']].dropna()

# Define the independent variables (add a constant term for the intercept)
new_X = new_model_data[['인센티브 변화', '지급한도 변화']]
new_X = sm.add_constant(new_X)

# Define the dependent variable
new_y = new_model_data['전달 대비 충전액']

# Create the regression model
new_model = sm.OLS(new_y, new_X).fit()

# Output the summary of the model
new_model.summary()

```
