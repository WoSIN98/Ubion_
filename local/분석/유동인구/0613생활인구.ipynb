```python
import pandas as pd 
import os 
```

```python
life_popul = pd.read_csv('../../data/경기도/원본/생활인구/경기도 가평군 생활인구 수_21년 12월.csv')
```

```python
## 31개 시군 csv 하나로 합치기

# 파일 경로 지정
directory_path = '../../data/경기도/원본/생활인구/'

# 모든 CSV 파일 목록 생성
all_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if f.endswith('.CSV')]

# 각 파일을 데이터프레임으로 읽어오고 하나의 리스트에 저장
dataframes = [pd.read_csv(file) for file in all_files]

# 모든 데이터프레임을 하나로 합치기
combined_csv = pd.concat(dataframes, ignore_index=True)
```

```python
combined_csv
```

```python
combined_csv.to_csv('../../data/경기도/원본/생활인구/전처리/31시군총합.csv',index=False)
```

```python
totallife_popul = combined_csv[['CRTR_YMD','DWK_NM','SGG_NM','TMZN_CD','REVISN_LIVE_PUL_CNT']]
totallife_popul
```

```python
totallife_popul.columns = ['날짜','요일','시군명','시간대','인구수']
totallife_popul
```

```python
## '날짜', '요일', '시군명', '시간대'를 기준으로 그룹화하여 '인구수' 합계 계산
totallife_popul = totallife_popul.groupby(['날짜', '요일', '시군명', '시간대']).agg({'인구수': 'sum'}).reset_index()
totallife_popul
```

```python
totallife_popul.to_csv('../../data/경기도/원본/생활인구/전처리/total.csv',index=False)
```

```python
## 대략적인 인구수 추이 확인
totallife_popul.loc[(totallife_popul['날짜'] == 20211204) & (totallife_popul['시군명'] == '가평군')]

# 평일 주말 구분을 두려하였으나, 상관없이 소비가 제일 많을 것으로 예상되는 시간인 11,12,13,18,19,20,21 시간대를 사용하기로 결정
# 
```

```python
## 시간대 필터
popul1 = totallife_popul[totallife_popul['시간대'].isin((11,12,13,18,19,20,21))]
popul1
```

```python
# 원본 데이터에서 '시군명'별로 그룹화하고, 하루 인구수 평균 계산
popul2 = popul1.groupby('시군명')['인구수'].mean().reset_index()
popul2['인구수'] = popul2['인구수'].astype(int)
popul2
```

```python
popul2.to_csv('../../data/경기도/원본/생활인구/전처리/생활인구하루평균.csv',index=False)
```
