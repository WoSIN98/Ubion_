```python
import pandas as pd
```

```python
df = pd.read_csv('../../../data/경기도/전처리 후/total_df/2023년 지역화폐,인구.csv')
df
```

```python
df[df['시군명'] == '고양시']
```

```python
df2 = df
```

```python
### 스케일링
df2['월별사용액비중'] = df2['월별사용액'].map(lambda x : x/sum(df2['월별사용액']))
```

```python
grouped = df2.groupby('시군명').agg({
    '신규가입자': 'sum',
    '월별충전액': 'sum',
    '월별사용액': 'sum',
    '총인구수': 'mean',
    '가맹점수': 'mean',
    '생활인구수': 'mean',
    '20~29세': 'mean',
    '30~39세': 'mean',
    '40~49세': 'mean',
    '50~59세': 'mean',
    '60세 이상': 'mean'
}).reset_index()
```
