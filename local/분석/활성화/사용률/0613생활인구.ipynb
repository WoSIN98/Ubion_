```python
import pandas as pd
```

```python
df = pd.read_csv('../../../data/경기도/전처리 후/회귀df결측치포함.csv')
life_popul = pd.read_csv('../../../data/경기도/원본/생활인구/전처리/생활인구하루평균.csv')
```

```python
test = pd.read_csv('../../../data/경기도/원본/.csv')
```

```python
df2 = df[['기준년월','시군명','총인구수','신규가입자수','월별충전액','월별사용액','가맹점수']]
df2
```

```python
df3 = pd.merge(df2, life_popul, how='left', on='시군명')
df3 = df3.rename(columns={'인구수' : '생활인구수'})
df3
```

```python
df3['인당사용액(생활)'] = (df3['월별사용액'] / df3['생활인구수']) * 1000000
df3['인당사용액(생활)'] = df3['인당사용액(생활)'].astype(int)
df3
```

```python
df3.to_csv('../../../data/경기도/전처리 후/생활인구포함.csv',index=False)
```
