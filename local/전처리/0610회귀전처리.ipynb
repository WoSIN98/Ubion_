```python
incen = pd.read_csv('../data/경기도/인센티브율/2023종합.csv')
```

```python
incen.rename(columns={'지역':'시군명'},inplace=True)
```

```python
## 인센티브 정보
incen
```

```python
incen.drop(['연간 지급한도', '시작일', '종료일'], axis=1, inplace=True)
```

```python
incen['포인트 지급한도']= incen['포인트 지급한도'].map(lambda x : x.lstrip('월').rstrip('원'). strip())
```

```python
incen.rename(columns={'포인트 지급한도':'월지급한도'},inplace=True)
```

```python
incen.to_csv('../data/경기도/전처리 후/incen.csv',index=False)
```

```python
df_combined = pd.merge(popul,pay, how='right', on=['기준년월','시군명'])
```

```python
df_combined.info()
```

```python
df_combined = pd.merge(df_combined,incen, how='left',on=['기준년월','시군명'])
```

```python
df_combined = pd.merge(df_combined,count_store, how='left', on='시군명')
```

```python
df_combined['인센티브'].unique()
```

```python
df_combined.info()
```

```python
## 인센티브율이 '-'인 행 제거
df_clean = df_combined.loc[df_combined['인센티브'] != '-']
```

```python
df_clean.info()
```

```python
df_clean.loc[df_clean['신규가입자수'].isna()]
```

```python
## 결측치 제거
df_clean = df_clean.loc[df_clean['신규가입자수'] != 0]
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean.info()
```

```python
# '인센티브'의 백분율 제거 및 숫자형으로 변환
df_clean['인센티브'] = df_clean['인센티브'].str.rstrip('%').astype(float) / 100
```

```python
# '월지급한도'의 쉼표 제거 및 숫자형으로 변환
df_clean['월지급한도'] = df_clean['월지급한도'].str.replace(',', '').astype(int)
```

```python
df_clean.head(4)
```

```python
df_clean.to_csv('../data/경기도/전처리 후/회귀df.csv',index=False)
```

```python
df_clean.info()
```

```python
df_clean.to_csv('../data/경기도/전처리 후/회귀df결측치포함.csv',index=False)
```
