```python
popul = pd.read_csv('./data/경기도/주민등록인구집계현황.csv', encoding='cp949')
```

```python
popul_2023 = popul.loc[popul['연도'] == 2023].loc[popul['월'] == 12].loc[popul['행정구역구분명'] == '시군']
```

```python
popul_2023.head(1)
```

```python
popul_2023 = popul_2023[['행정구역명','총 인구수']].reset_index(drop=True)
```

```python
popul_2023
```

```python
popul_2023['행정구역명'] = popul_2023['행정구역명'].map(lambda x: x.strip('경기도'))
```

```python
popul_2023['행정구역명'] = popul_2023['행정구역명'].map(lambda x: x.strip())
```

```python
popul_2023.columns = ['시군명','총인구수']
```

```python
# 2023년 경기도 인구
popul_2023 
```

```python
popul_2023.to_csv('./data/경기도/2023년경기도인구.csv',index=False)
```

```python
total = pd.merge(total_update,popul_2023, how='outer', on='시군명')
```

```python
total['인구수대비사용액'] = total['총사용액'] / total['총인구수']
```

```python
total.sort_values('인구수대비사용액',ascending=False,inplace=True)
```

```python
total.reset_index(drop=True,inplace=True)
```

```python
total.to_csv('./data/경기도/2023년인구수대비사용액.csv',index=False)
```
